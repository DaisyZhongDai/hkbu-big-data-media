### Data Analysis

```python
df=pd.read_csv('Attachment A of gender recognition paper published by HKSAR government.csv')
```

```python
df[:3]
```

```python
df['Real life test'].value_counts()
```

```python
df['Real life test'] = df['Real life test'].replace('\n', 'no')
df['Real life test'] = df['Real life test'].replace('(2 yrs)', 'yes')
df['Real life test'] = df['Real life test'].replace('✗', 'no')
df['Real life test'] = df['Real life test'].replace('✗\n', 'no')

def check_gou1(x):
    if x is not None:
        x=str(x)
        if '✓' in x:
            df['Real life test'] = df['Real life test'].replace(x, 'yes')
df['Real life test'].apply(check_gou1)

def check_unclear1(x):
    if x is not None:
        x=str(x)
        if 'Unclear' in x:
            df['Real life test'] = df['Real life test'].replace(x, 'unclear')
            
df['Real life test'].apply(check_unclear1)
```

```python
df['Medical diagnosis e.g. gender dysphoria, GID etc'].value_counts()
```

```python
df['Medical diagnosis e.g. gender dysphoria, GID etc'] = df['Medical diagnosis e.g. gender dysphoria, GID etc'].replace('✗', 'no')
df['Medical diagnosis e.g. gender dysphoria, GID etc'] = df['Medical diagnosis e.g. gender dysphoria, GID etc'].replace("✗\n", 'no')
def check_gou(x):
    if x is not None:
        x=str(x)
        if '✓' in x:
            df['Medical diagnosis e.g. gender dysphoria, GID etc'] = df['Medical diagnosis e.g. gender dysphoria, GID etc'].replace(x, 'yes')
            
df['Medical diagnosis e.g. gender dysphoria, GID etc'].apply(check_gou)

def check_unclear(x):
    if x is not None:
        x=str(x)
        if 'Unclear' in x:
            df['Medical diagnosis e.g. gender dysphoria, GID etc'] = df['Medical diagnosis e.g. gender dysphoria, GID etc'].replace(x, 'unclear')
            
df['Medical diagnosis e.g. gender dysphoria, GID etc'].apply(check_unclear)
```

```python
df['Physical adaptation (incl non-genital surgery)'].value_counts()
```

```python
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('✗\n', 'no') 
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('✗', 'no')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('\n', 'no')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n (surgical, hormonal or other treatment appropriate for the purpose of gender transition)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('(sex reassignment procedure required, extent unclear)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(surgical, hormonal or other treatment appropriate for that individual for the purpose of gender transition)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear(surgical, hormonal or other treatment appropriate for gender transition or intersex condition)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(alteration of sexual characteristics by surgical or medical treatment)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(requested sex designation more accurately reflects applicant’s sex or gender identity)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(appropriate clinical treatment required)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('✓\n', 'yes')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('✓\n(has gender characteristics of the sex reassigned)\n', 'yes')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(“sexual reassignment” & “physical conformation” to be determined by medical experts)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(medical internvention, endocrinolo-gical therapy)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('Unclear\n(modifying sexual characteristics needed)', 'unclear')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace('✗\n(surgical and/or hormone treatment or other treatment appropriate for the individual)', 'no')
```

```python
df.to_csv('Attachment A_clean.csv')
```

```python
del df['Parental status exclusion']
del df['Official documents etc, affected']
del df['Recognition of foreign GRS/ medical intervention/ foreign gender recognition']
df['Genital surgery leading to sterilization'] = df['Genital surgery leading to sterilization'].replace(np.nan, 'no')
df['Physical adaptation (incl non-genital surgery)'] = df['Physical adaptation (incl non-genital surgery)'].replace(np.nan, 'no')
df['hormonal treatment'] = df['hormonal treatment'].replace(np.nan, 'no')
df['Real life test'] = df['Real life test'].replace(np.nan, 'no')
df['Medical diagnosis e.g. gender dysphoria, GID etc'] = df['Medical diagnosis e.g. gender dysphoria, GID etc'].replace(np.nan, 'no')
```

```python
df['Genital surgery leading to sterilization'] = df['Genital surgery leading to sterilization'].replace('(sex reassignment procedure required, extent unclear)', 'unclear')

def check_unclear1(x):
    if x is not None:
        x=str(x)
        if 'Unclear' in x:
            df['Genital surgery leading to sterilization'] = df['Genital surgery leading to sterilization'].replace(x, 'unclear')
df['Genital surgery leading to sterilization'].apply(check_unclear1)

def check_gou1(x):
    if x is not None:
        x=str(x)
        if '✓' in x:
            df['Genital surgery leading to sterilization'] = df['Genital surgery leading to sterilization'].replace(x, 'yes')
df['Genital surgery leading to sterilization'].apply(check_gou1)

def check_cha1(x):
    if x is not None:
        x=str(x)
        if '✗' in x:
            df['Genital surgery leading to sterilization'] = df['Genital surgery leading to sterilization'].replace(x, 'no')
df['Genital surgery leading to sterilization'].apply(check_cha1)
```

```python
def check_cha(x):
    if x is not None:
        x=str(x)
        if '✗' in x:
            df['hormonal treatment'] = df['hormonal treatment'].replace(x, 'no')
df['hormonal treatment'].apply(check_cha)
df['hormonal treatment'] = df['hormonal treatment'].replace('\n', 'no')
def check_unclear(x):
    if x is not None:
        x=str(x)
        if 'Unclear' in x:
            df['hormonal treatment'] = df['hormonal treatment'].replace(x, 'unclear')
            
df['hormonal treatment'].apply(check_unclear)
def check_gou(x):
    if x is not None:
        x=str(x)
        if '✓' in x:
            df['hormonal treatment'] = df['hormonal treatment'].replace(x, 'yes')
            
df['hormonal treatment'].apply(check_gou)
```
