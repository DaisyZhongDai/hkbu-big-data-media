```python
 import requests
import bs4
import csv
```

```python
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://www.qidian.com/all')
```

```python
def get_the_page(url):
    r=requests.get(url)
    mypage=BeautifulSoup(r.text)
    information=mypage.find_all('div', attrs={'class': 'book-mid-info'})
    name=[]
    writer=[]
    genre=[]
    genre_sub=[]
    intro=[]
    words=[]
    for i in range(0,20):
        name.append(information[i].find('h4').text.strip())
        writer.append(information[i].find('a', attrs={'class': 'name'}).text.strip())
        genre1=information[i].find('p', attrs={'class':'author'})
        genre_sub.append(genre1.find('a', attrs={'class':'go-sub-type'}).text.strip())
        genre.append(genre1.find_all('a')[1].text.strip())
        intro.append(information[i].find('p', attrs={'class':'intro'}).text.strip())
        words.append(information[i].find('p', attrs={'class':'update'}).text.strip())

    my_first_page=[]
    for i in range(0,20):
        my_first_page.append([name[i],writer[i],genre[i],genre_sub[i],intro[i],words[i]])
    return my_first_page

```

```python
pages=[]
for q in range(1,11):
    pagenumber=str(q)
    myurl="https://www.qidian.com/all?orderId=&style=1&pageSize=20&siteid=1&pubflag=0&hiddenField=0&page="+pagenumber
    pages.extend(get_the_page(myurl))

pages
```

```python
import csv
with open('daisy.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerows(pages)
```
